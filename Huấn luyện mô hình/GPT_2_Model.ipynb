```python
from datasets import load_dataset, Dataset, DatasetDict
from transformers import GPT2Tokenizer

# Load the dataset
dataset = load_dataset("kienhoang123/VN_legal_framewok")

# Create a validation split from the training data
train_val = dataset['train'].train_test_split(test_size=0.1, seed=42)

# Create a new dataset dictionary with both splits
dataset = DatasetDict({
    'train': train_val['train'],
    'validation': train_val['test']
})

# Print column names to identify the text column
print("Available columns:", dataset['train'].column_names)

# Load GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token

# Tokenization function
def tokenize_function(examples):
    return tokenizer(
        examples['nội_dung'],  # Changed to use the correct column name
        truncation=True,
        max_length=512,
        padding='max_length',
        return_tensors='pt'
    )

# Tokenize the datasets
tokenized_train = dataset['train'].map(
    tokenize_function,
    batched=True,
    remove_columns=dataset['train'].column_names
)
tokenized_val = dataset['validation'].map(
    tokenize_function,
    batched=True,
    remove_columns=dataset['validation'].column_names
)

# Print tokenized dataset info
print("\nTokenized dataset information:")
print(f"Training set size: {len(tokenized_train)}")
print(f"Validation set size: {len(tokenized_val)}")
print("\nExample of tokenized output:")
print(tokenized_train[0])
```

```python
from datasets import load_dataset, Dataset, DatasetDict
from transformers import AutoTokenizer, DataCollatorForLanguageModeling

# ... previous dataset loading and tokenization code ...

# Setup the data collator for language modeling
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # We're doing causal language modeling, not masked language modeling
)

# Test the data collator with a small batch
batch_size = 4
sample_batch = {
    key: tokenized_train[key][:batch_size]
    for key in tokenized_train.features
}

# Create a batch using the data collator
collated_batch = data_collator([dict(zip(sample_batch, t)) for t in zip(*sample_batch.values())])

# Print information about the collated batch
print("\nData Collator Output Example:")
print("Input shape:", collated_batch['input_ids'].shape)
print("Labels shape:", collated_batch['labels'].shape)
print("\nFirst sequence input_ids:")
print(collated_batch['input_ids'][0])
print("\nCorresponding labels (shifted right by one position):")
print(collated_batch['labels'][0])
```

```python
from transformers import Trainer, TrainingArguments, GPT2LMHeadModel

# Initialize the model
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Define training arguments
training_args = TrainingArguments(
    output_dir="/Users/Kien/Documents/RAG/results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir="/Users/Kien/Documents/RAG/logs",
    logging_steps=100,
    save_steps=1000,
    eval_steps=1000,
    evaluation_strategy="steps",
    save_total_limit=3,
    load_best_model_at_end=True,
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    data_collator=data_collator,
)

# Start training
print("Starting training...")
trainer.train()

# Save the final model
trainer.save_model("/Users/Kien/Documents/RAG/final_model")
print("Training completed and model saved!")
```

```python
import math
from transformers import pipeline

# Load the fine-tuned model and tokenizer
generator = pipeline('text-generation',
                    model='/Users/Kien/Documents/RAG/final_model',
                    tokenizer=tokenizer)

# Test prompts (you can modify these with relevant legal text prompts in Vietnamese)
test_prompts = [
    "Điều 1. Phạm vi điều chỉnh",
    "Quyền và nghĩa vụ của",
    "Theo quy định của pháp luật"
]

# Generate text for each prompt
print("Generated Text Examples:")
for prompt in test_prompts:
    print("\nPrompt:", prompt)
    generated_text = generator(
        prompt,
        max_length=200,
        num_return_sequences=1,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id
    )
    print("Generated:", generated_text[0]['generated_text'])

# Calculate perplexity on validation set
eval_results = trainer.evaluate()
print("\nValidation Results:")
print(f"Perplexity: {math.exp(eval_results['eval_loss']):.2f}")
```
